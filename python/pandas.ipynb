```python
import pandas as pd
```

```python
# DataFrame - coverts given  data into table format
# for a single list of data ,it creates a table with 'n' no.rows equal to the no. of elements in the list

dict1 = {
   " name ":['Sana','Samreen','Ayesha','Asma'],
   " marks" : [90,92,95,98],
   "city": ['hyd','vizag','Nizambad','soudi']
}

df=pd.DataFrame(dict1)
print(df)
# to_csv is a function used to to create a csv file in a excel sheet
df.to_csv('friends_data.csv', index=False)
# pd.read_csv(dict1) - reads data from a csv file
```

```python
#  creating a dataframe containing random data

import numpy as np

df = pd.DataFrame(np.random.rand(100,5))
print(df)
# head()- prints first 5 rows of the DataFrame
print(df.head())
# tail()- prints last 5 rows of the DataFrame
print(df.tail())
type(df)
```

```python
# series - it is defined as a single column or a 1-D list , it can contain data of different datatypes

s = pd.Series([1, 3, 5, np.nan, 6, 8,"sa"])
print(s)
```

```python
# modifying column names

table = pd.DataFrame(np.random.rand(5,5))
table
table.columns=["A","B","C","D","E"]
table
```

```python
# modifying values in dataframe using "loc"
new_table = table.copy()
new_table.loc[0,"A"] = 1234
new_table
new_table.pop(1)
new_table
```

```python
# acessing rows and columns in a table using "iloc"
table = pd.DataFrame(np.random.rand(5,5))
newTable2 = table.copy()
newTable2.iloc[0,3]
print(newTable2)
```

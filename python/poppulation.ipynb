```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('world_population_data.csv')
data
```

```python
sorted_table = data.groupby('continent')['2023 population'].sum().reset_index(name="population")

population = sorted_table['population']
continent = sorted_table['continent']
population.to_list()
continent.to_list()
plt.bar(np.arange(len(population)), population,  color='skyblue', label="Population")
plt.xticks(np.arange(len(population)), continent, rotation = 45)

# Set labels and title
plt.xlabel('Population')
plt.ylabel('Continent')
plt.title('Population by Continent in 2023')

# Show the plot
plt.show()
```

```python
# population growth by time series analysis 
# populations = data[['2023 population', '2022 population', '2020 population','2020 population','2015 population','2010 population','2000 population','1990 population','1980 population','1970 population']]
population_columns = ['2023 population', '2022 population', '2020 population', '2015 population', '2010 population', '2000 population', '1990 population', '1980 population', '1970 population']

# Group by 'continent' and sum up the populations for all years
data2 = data.groupby('continent')[population_columns].sum().transpose()

# Plotting
data2.plot(kind='line', marker='o', figsize=(12, 8))
plt.title('Population Growth by Continent Over the Years')
plt.xlabel('Year')

plt.ylabel('Population')
plt.legend(title='Continent', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
# data2 = data.groupby(['continent','country'])['populations'].sum().reset_index()
```

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('jobs_in_data.csv')
data
```

```python
# jobs trend by jobs category

job_category_column = data['job_category']
year_column = data['work_year']
my_data = pd.DataFrame({'Year':year_column, 'Job Category':job_category_column})
my_data

Jobs = my_data['Job Category']
Years = my_data['Year']

job_counts = my_data.groupby(['Year', 'Job Category']).size().reset_index(name='count')
print(job_counts)
# Plot the graph
plt.figure(figsize=(10, 6))

# Iterate through unique job categories
for job_category in job_counts['Job Category'].unique():
    subset = job_counts[job_counts['Job Category'] == job_category]
    plt.plot(subset['Year'], subset['count'],marker = 'o', label=job_category)

plt.title('Job Counts by Category and Year')
plt.xlabel('Year')
plt.ylabel('Job Count')
plt.legend()
plt.show()

```

```python
# jobs count by jobs category

job_category_column = data['job_category']
category_count = job_category_column.value_counts()

Jobs_count_list = []
Category_list = []

for job, category in category_count.items():
    Jobs_count_list.append(job)
    Category_list.append(category)

print(Jobs_count_list)
print(Category_list)

x_axis = range(0,3000,300)
y_axis = Jobs_count_list
plt.xlabel('category')
plt.ylabel('Count')
plt.title('Jobs Count by Jobs Category')
color = ['blue', 'orange', 'green','red', 'violet','brown', 'pink','yellow', 'black', 'gray']
y_axis.reverse()
plt.barh(y_axis, x_axis, label = Category_list, color = color)
```
